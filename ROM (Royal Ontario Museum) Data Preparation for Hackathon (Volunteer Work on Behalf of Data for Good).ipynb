### Royal Ontario Museum (ROM) is Toronto's premire cultural institution. They plan to analyze their membership database to enhance marketing campaign and rely on Data for Good (Toronto chapter of Not-for-Profit orgaization) to prepare the data for the upcoming hackathon. I volunteered for data cleaning session held on July 28 2018. Here's my contribution.

```python
import pandas as pd
```

```python
# There are 7 datasets. Due to the large size, each dataset is stored in 10 different databases.
# They are needed to be merged together before analysis. These datasets are taken from Relational database.
# Primary key is 'MemCnBio_Import_ID' in memCnBio table.

Mem = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_.txt', sep='\t', index_col=None)
    Mem = pd.concat([Mem,x], ignore_index=True)
    
MemArt = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_2.txt', sep='\t', index_col=None)
    MemArt = pd.concat([MemArt,x], ignore_index=True)

MemArtCat_1 = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_3.txt', sep='\t', index_col=None)
    MemArtCat_1 = pd.concat([MemArtCat_1,x], ignore_index=True)

MemCn = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_4.txt', sep='\t', index_col=None)
    MemCn = pd.concat([MemCn,x], ignore_index=True)
    
MemCnBio = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_5.txt', sep='\t', index_col=None)
    MemCnBio = pd.concat([MemCnBio,x], ignore_index=True)

MemHst_1 = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_6.txt', sep='\t', index_col=None)
    MemHst_1 = pd.concat([MemHst_1,x], ignore_index=True)

MemHst_1Gft_1  = pd.DataFrame()
for i in range(0,11):
    x = pd.read_csv('ROM_db' + str(i) + '_7.txt', sep='\t', index_col=None)
    MemHst_1Gft_1 = pd.concat([MemHst_1Gft_1,x], ignore_index=True)
```

```python
# Although primary key should be unique, there are many duplication in MemCnBio key
MemCnBio.groupby('MemCnBio_Import_ID')['MemCnBio_Import_ID', 'database_num'].count()[:10]
```

```python
# Create the table of duplicated MemCnBio_Import_ID
duplicate = MemCnBio.groupby('MemCnBio_Import_ID')[['MemCnBio_Import_ID']].count() 
sorted_dup = duplicate.sort_values('MemCnBio_Import_ID', ascending=False)
```

```python
sorted_dup[sorted_dup['MemCnBio_Import_ID']>1]
```

```python
# Extract idex values, which are MemCnBio_Import_IDs appearing in the records more than 2 times, to be used for filter
duplicated_ID = sorted_dup[sorted_dup['MemCnBio_Import_ID']>1].index.values
duplicated_ID
```

```python
# There are 36059 records with duplicate MemCnBio_Import_ID, which is about right number accorind to the previous check
y = 0
for x in range(0, len(MemCnBio)):
     y += MemCnBio['MemCnBio_Import_ID'].iloc[x] in duplicated_ID
        
print(y)
```

```python
# Create True/False column to show if MemCnBio_Import_ID is duplicated.
MemCnBio['duplicate'] = MemCnBio['MemCnBio_Import_ID'].apply(lambda x: x in duplicated_ID)
```
